## Análise dos dados

```python
# Tabela organizada por IDH por Estado

file_idh = file.sort_values(by='IDH_2010')
file_idh
```

```python
# Tabela organizada por Receitas Realizadas por Estado

file_receitas = file[['UF','ReceitasRealizadas_2017']].sort_values(by='ReceitasRealizadas_2017')
file_receitas
```

```python
# Tabela organizada por Despesas Empenhadas por Estado
file_despesas = file.sort_values(by='DespesasEmpenhadas_2017')
file_despesas
```

## Plotagem de gráficos

```python
# Gráfico de IDH por Estado

ax = sns.barplot(x='UF', y='IDH_2010', data=file_idh)
ax.set_xticklabels(ax.get_xticklabels(), rotation=50, ha="right")
ax
```

```python
#Boxplot de IDH por Estado

df_idh = pd.DataFrame(file_idh['IDH_2010'])
sns.boxplot(x="IDH_2010", data=file_idh)
```

```python
# Gráfico de Receitas Realizadas por Estado

ax_rec = sns.barplot(x='UF', y='ReceitasRealizadas_2017', data=file_receitas)
ax_rec.set_xticklabels(ax_rec.get_xticklabels(), rotation=50, ha="right")
```

```python
# Gráfico de Despesas Empenhadas por Estado 

ax_rec = sns.barplot(x='UF', y='DespesasEmpenhadas_2017', data=file_despesas)
ax_rec.set_xticklabels(ax.get_xticklabels(), rotation=50, ha="right")
ax
```
